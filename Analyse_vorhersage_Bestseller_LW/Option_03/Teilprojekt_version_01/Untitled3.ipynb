```python
from bs4 import BeautifulSoup
import pandas as pd

# Beide HTML-Blöcke in einem String
html = """
<h1>APP.3.2.A20 Benennung von Anzusprechenden (S) [Zentrale Verwaltung]</h1>
<p>Bei umfangreichen Webangeboten SOLLTE die Institution zentrale Anzusprechende für die Webangebote bestimmen. Es SOLLTEN Prozesse, Vorgehensweisen und Zuständige für Probleme oder Sicherheitsvorfälle benannt werden.</p>
<p>Die Institution SOLLTE eine Kontaktmöglichkeit auf ihrer Webseite veröffentlichen, über die Sicherheitsprobleme an die Institution gemeldet werden können. Für die Behandlung von externen Sicherheitsmeldungen SOLLTE die Institution Prozesse definieren.</p>

<h1>APP.6.A1 Planung des Software-Einsatzes (B) [Fachverantwortliche]</h1>
<p>Bevor eine Institution eine (neue) Software einführt, MUSS sie entscheiden,</p>
<ul>
<li><p>wofür die Software genutzt und welche Informationen damit verarbeitet werden sollen,</p></li>
<li><p>wie die Benutzenden bei der Anforderungserhebung beteiligt und bei der Einführung unterstützt werden sollen,</p></li>
<li><p>wie die Software an weitere Anwendungen und IT-Systeme über welche Schnittstellen angebunden wird,</p></li>
<li><p>auf welchen IT-Systemen die Software ausgeführt werden soll und welche Ressourcen zur Ausführung der Software erforderlich sind, sowie</p></li>
<li><p>ob sich die Institution in Abhängigkeit zu einem Hersteller oder einer Herstellerin begibt, wenn sie diese Software einsetzt.</p></li>
</ul>
<p>Hierbei MÜSSEN bereits Sicherheitsaspekte berücksichtigt werden. Zusätzlich MUSS die Institution die Zuständigkeiten für fachliche Betreuung, Freigabe und betriebliche Administration schon im Vorfeld klären und festlegen. Die Zuständigkeiten MÜSSEN dokumentiert und bei Bedarf aktualisiert werden.</p>
"""

# HTML parsen
soup = BeautifulSoup(html, 'html.parser')

data = []

# Durchlaufe alle H1-Blöcke
for h1 in soup.find_all('h1'):
    nummer, titel = h1.text.strip().split(' ', 1)

    # Sammle alle folgenden <p>, <ul>, <li> etc. bis zum nächsten <h1>
    content = []
    sibling = h1.find_next_sibling()
    while sibling and sibling.name != 'h1':
        content.append(sibling.get_text(strip=True))
        sibling = sibling.find_next_sibling()

    full_text = ' '.join(content)
    data.append({
        "Nummer": nummer,
        "Titel": titel,
        "Inhalt": full_text
    })

# In DataFrame umwandeln
df = pd.DataFrame(data)

# Ergebnis anzeigen
df

# Optional: in Excel oder CSV speichern
# df.to_csv("anforderungen.csv", index=False)

```
